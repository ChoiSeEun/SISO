## 구별 노령화 지수

```python
pop = pd.read_csv('data/서울시 행정동 단위 거주인구 데이터.txt', encoding='utf-8', sep='|')
pop.head()
```

```python
df_pop = pop.loc[pop['STD_YM'] == 202106]
df_pop.info()
```

```python
df_pop['AGRDE_CD'].unique()
```

```python
df = df_pop.copy()

df.loc[df['AGRDE_CD']<'1014', 'AGE'] = '10세 미만'
df.loc[(df['AGRDE_CD']>'0509') & (df['AGRDE_CD']<'2024'), 'AGRP'] = '10대'
df.loc[(df['AGRDE_CD']>'1519') & (df['AGRDE_CD']<'3034'), 'AGRP'] = '20대'
df.loc[(df['AGRDE_CD']>'2529') & (df['AGRDE_CD']<'4044'), 'AGRP'] = '30대'
df.loc[(df['AGRDE_CD']>'3539') & (df['AGRDE_CD']<'5054'), 'AGRP'] = '40대'
df.loc[(df['AGRDE_CD']>'4549') & (df['AGRDE_CD']<'6064'), 'AGRP'] = '50대'
df.loc[(df['AGRDE_CD']>'5559') & (df['AGRDE_CD']<'7074'), 'AGRP'] = '60대'
df.loc[(df['AGRDE_CD']>'6569') & (df['AGRDE_CD']<'8084'), 'AGRP'] = '70대'
df.loc[(df['AGRDE_CD']>'7579') & (df['AGRDE_CD']<'9094'), 'AGRP'] = '80대'
df.loc[(df['AGRDE_CD']>'8589') & (df['AGRDE_CD']<'I00'), 'AGRP'] = '90대'
df.loc[df['AGRDE_CD']>'9599', 'AGRP'] = '100세 이상'

df.loc[df_pop['AGRDE_CD'] > '6064', 'AGRP'] = '65세 이상'
df.loc[df_pop['AGRDE_CD'] < '1519', 'AGRP'] = '15세 미만'

df['AGRP'].unique()
```

```python
gu = df.groupby(['CTY_NM', 'AGRP']).sum()['RSPOP_CNT'].reset_index(name='cnt')
gu
```

```python
lst = gu['CTY_NM'].unique().tolist()
gu_15 = gu.loc[gu['AGRP'] == '15세 미만'][['cnt']]
gu_15 = list(gu_15['cnt'])
gu_65 = gu.loc[gu['AGRP'] == '65세 이상'][['cnt']]
gu_65 = list(gu_65['cnt'])
```

```python
df_ratio = pd.DataFrame({'CTY_NM':lst, '15세 미만':gu_15, '65세 이상':gu_65})
df_ratio['ratio'] = df_ratio['65세 이상'] / df_ratio['15세 미만'] * 100
df_ratio
```

## 구별 학교 수

```python
sch = pd.read_csv('data/서울시학교.csv')
sch.head()
```

```python
sch.info()
```

```python
sch['설립구분'].unique()
```

```python
df_s = sch.loc[sch['설립구분'] != '사립']
df_s['설립구분'].unique()
```

```python
gu_sch = df_s['구'].value_counts().reset_index(name='school')
#gu_sch = gu_sch.sort_values()
df_sch= gu_sch.sort_values(by='index')
df_sch = df_sch.rename(columns={'index' : 'CTY_NM'})
df_sch['CTY_NM'] = df_sch['CTY_NM'].str.strip()
df_sch
```

## 구별 보건소 수

```python
hos = pd.read_csv('data/서울시 보건소.csv', encoding='utf-8')
hos.head()
```

```python
hos['hos'] = hos.sum(axis=1)
hos.head()
```

```python
df_hos = hos.groupby(['자치구']).sum()['hos'].reset_index()
idx = df_hos[df_hos['자치구'] == '합계'].index
df_hos = df_hos.drop(idx)
df_hos = df_hos.rename(columns={'자치구' : 'CTY_NM'})
df_hos['CTY_NM'] = df_hos['CTY_NM'].str.strip()
df_hos
```

## 구별 복지관 수

### 사회복지시설 목록 데이터

```python
df = pd.read_csv('data/서울특별시 사회복지시설 목록.csv', encoding='euc-kr')
df
```

```python
df.info()
```

```python
sen = ['(노인) 노인요양시설', '(노인) 양로시설', '(노인) 노인복지관(소규모)','(노인) 노인보호전문기관', '(노인) 노인복지관', 
       '(노인) 노인일자리지원기관','(노인) 노인복지주택', '(노인) 노인공동생활가정', '(노인) 어르신상담센터', '(노인) 노인교실', 
       '(노인) (기타)노인복지시설', '(노인) 경로식당', '(노인) 노인복지센터']
```

```python
df_bok = df.loc[df['시설종류명(시설유형)'].isin(sen)]
df_bok = df_bok['시군구명'].value_counts().reset_index(name='bok')
df_bok = df_bok.drop([df_bok.index[25]])
df_bok = df_bok.sort_values(by='index')
df_bok = df_bok.rename(columns={'index' : 'CTY_NM'})
df_bok
```

### 경로당 데이터

```python
df = pd.read_csv('data/서울시 노인여가 복지시설 (구별) 통계.csv')
df = df.drop([df.index[0], df.index[1]])
df = df[['자치구', '경로당']]
df_2 = df.sort_values(by='자치구')
df_2 = df_2.rename(columns={'자치구' : 'CTY_NM'})
df_2['경로당'] = pd.to_numeric(df['경로당'])
df_2.info()
```

### 복지관 + 경로당

```python
df_center = pd.merge(df_bok, df_2, on='CTY_NM')
df_center['center'] = df_center['bok'] + df_center['경로당']
df_center = df_center[['CTY_NM', 'center']]
df_center.head()
```

## 구별 문화센터 수

```python
culture = pd.read_csv('data/서울시 문화시설 현황.csv', encoding='euc-kr')
culture.head()
```

```python
culture.info()
```

```python
c = ['문화센터', '기타 문화시설', '문화회관']
culture = culture.loc[culture['분류3'].isin(c)]
culture.head()
```

```python
df_culture = culture['행정 구'].value_counts().reset_index(name='culture')
df_culture = df_culture.sort_values(by='index')
df_culture = df_culture.rename(columns={'index' : 'CTY_NM'})
df_culture['CTY_NM'] = df_culture['CTY_NM'].str.strip()
df_culture
```

## 최종 분석 파일 (구 클러스터링에 사용)

```python
df_old = df_ratio[['CTY_NM', 'ratio']]
df_old.head()
```

```python
df_sch = df_sch.reset_index()
df_sch = df_sch.drop(['index'], axis='columns')
df_sch.head()
```

```python
df_hos.head()
```

```python
df_culture = df_culture.reset_index()
df_culture = df_culture.drop(['index'], axis='columns')
df_culture.head()
```

```python
df_center.head()
```

```python
df = pd.DataFrame({'CTY_NM':lst})
df = pd.merge(df, df_old, on='CTY_NM', how='inner')
df = pd.merge(df, df_sch, how='inner', on='CTY_NM')
df = pd.merge(df, df_hos, how='inner', on='CTY_NM')
df = pd.merge(df, df_culture, how='outer', on='CTY_NM')
df = pd.merge(df, df_center, how='inner', on='CTY_NM')
df = df.fillna(0)
df = df.astype({'culture':int})
df['tot'] = df['school'] + df['hos'] + df['culture'] + df['center'] 
df = df[['CTY_NM', 'ratio', 'tot']]
df
```
