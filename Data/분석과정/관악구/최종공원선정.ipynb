# 공원 선정

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('관악구_시니어파크_후보군.csv',encoding = 'cp949')
data.head()
```

## 불필요한 컬럼제거

```python
data.drop(['공원보유_1','공원보유시','공원보유_2','관리기관명','전화번호','데이터기준','공원보유_3','공원보유_4','지정고시일','소재지지번'],axis = 1,inplace = True)
data.head()
```

```python
data.columns = ['관리번호', '공원명', '공원구분', '소재지도로', '위도', '경도', '공원면적', 'NUM_고령인구', 'NUM_전통시장', 'NUM_주민센터', 'NUM_복지시설', 'NUM_자전거', 'NUM_지하철', 'NUM_버스정류장']
data.head()
```

```python
data['INDEX_대중교통'] = data['NUM_자전거']*(0.9/1.4) + data['NUM_지하철']*(0.4/1.4) + data['NUM_버스정류장']*(0.1/1.4)
data.drop(['NUM_자전거','NUM_지하철','NUM_버스정류장'],axis = 1,inplace = True)
data.head()
```

```python
min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()
```

```python
min_max_scaler.fit(data[['NUM_고령인구','NUM_전통시장','NUM_주민센터','NUM_복지시설','INDEX_대중교통']])
score_array = min_max_scaler.transform(data[['NUM_고령인구','NUM_전통시장','NUM_주민센터','NUM_복지시설','INDEX_대중교통']])
```

```python
data2 = pd.DataFrame(score_array,columns = ['인구점수','전통시장점수','주민센터점수','복지시설점수','대중교통점수'])
data2.head()
```

```python
final_data = pd.concat([data,data2],axis = 1)
final_data.head()
```

---
### standard scaler

```python
standard_scaler.fit(data[['NUM_고령인구','NUM_전통시장','NUM_주민센터','NUM_복지시설','INDEX_대중교통']])
score_array2 = standard_scaler.transform(data[['NUM_고령인구','NUM_전통시장','NUM_주민센터','NUM_복지시설','INDEX_대중교통']])
```

```python
data3 = pd.DataFrame(score_array2,columns = ['인구점수','전통시장점수','주민센터점수','복지시설점수','대중교통점수'])
data3.head()
```

```python
final_data = pd.concat([data,data2],axis = 1)
final_data.head()
```

## 점수합산

```python
final_data['최종점수'] = final_data['인구점수'] + final_data['전통시장점수'] + final_data['주민센터점수'] + final_data['복지시설점수'] + final_data['대중교통점수']
final_data.sort_values(by = '최종점수', ascending = False)
```
